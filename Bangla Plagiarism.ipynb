```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Plagiarism Detection/pre-processed-bangla-15K.csv')
data.reset_index(inplace=True)
data.head()
```

```python
data.drop(['index','Unnamed: 0'],inplace=True, axis=1)
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0,max_features=5000)
tfidf_matrix = tf.fit_transform(data['text'])
```

```python
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
print(len(cosine_similarities))
```

```python
data.insert(0, 'uniq_id', range(1, 1 + len(data)))
data.head()
```

```python

def result():
  results = {}

  for idx, row in data.iterrows():
      similar_indices = cosine_similarities[idx].argsort()[:-100:-1]
      similar_items = [(cosine_similarities[idx][i], data['uniq_id'][i]) for i in similar_indices]

      results[row['uniq_id']] = similar_items[1:]
      
  return results
results = result()
```

```python
def item(id):
    return data.loc[data['uniq_id'] == id]['text'].tolist()[0].split(' - ')[0]
```

```python
# Just reads the results out of the dictionary.
def recommend(item_id, num):
    print("Showing " + str(num) + " contents similar to " + item(item_id)[:50] + "...")
    print("-------")
    recs = results[item_id][:num]
    for rec in recs:
        print("Plagiarised: " + item(rec[1])[:50] + " (score:" + str(round(rec[0]*100,2) ) + ")%"[:50])

recommend(item_id=4, num=5)
```

```python
def preprocess(text):
  per_text = ''
  processed_text = ''
  for ch in text:
    if ch not in punctuations:
      per_text+=ch
      # print(per_text)
  for word in per_text:
    if word not in bn_stop_words:
      processed_text+=word
      # print(processed_text)
  processed_text.replace('\n',' ')
  return processed_text
```
